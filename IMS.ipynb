Add New Item into Inventory

```python
n = int(input("Enter no. of products to be added:"))
for i in range(n):
  prod_id = str(input("Enter product id:"))
  if prod_id in dct.keys(): 
    print("ID already taken, please enter valid key")
  else:
    name = str(input("Enter name:"))
    pr = int(input("Enter price:"))
    qn = int(input("Enter quantity:"))
    dis = int(input("Enter Discount:"))
    cg = str(input("Enter Category:"))
  i = i - 1
record[prod_id] = {'name': name, 'price': pr, 'quantity': qn, 'Discount':dis,'category':cg}
```

```python
js = json.dumps(record)

fd = open("records.json",'w')
fd.write(js)
fd.close()
```


Purchase

```python
import time

total_prod = int(input("Enter total number of products to be added in cart:"))
dis = 0
total_bill = 0
print("*********************************************************")
for i in range(total_prod):
  ui_prod  = str(input("Enter the product_Id: "))
  ui_quant = int(input("Enter the quantity: "))
  print("Product: ", record[ui_prod]['name'])
  print("Price: ", record[ui_prod]['price'])
  print("Billing Amount: ", record[ui_prod]['price'] * ui_quant)
  print("Discount:", record[ui_prod]['discount']*ui_quant)
  total_bill =total_bill + record[ui_prod]['price'] * ui_quant
  dis = dis + record[ui_prod]['discount']*ui_quant
  i = i-1
print("*************************************")
print("Total Bill:",total_bill)
print("Total Discount:",dis)
print("*************************************")
print(time.ctime())
print("Thank you!!!")
record[ui_prod]['quantity'] = record[ui_prod]['quantity'] - ui_quant
```
